```python
import pandas as pd

df_index_lemmas = pd.read_parquet("/export/usuarios_ml4ds/lbartolome/NextProcurement/sproc/place_feb_21/place_all_embeddings_metadata.parquet")
##pd.read_parquet("/export/usuarios_ml4ds/lbartolome/NextProcurement/NP-Backend-Dockers/data/source/PLACE_all.parquet") 
df_info_objectives = pd.read_parquet("/export/usuarios_ml4ds/lbartolome/NextProcurement/NP-Text_Object/data/train_data/processed_objectives_with_lemmas_emb_label_tm_augmented.parquet")
```

```python
#df_index_lemmas["embeddings"] = df_index_lemmas["embeddings"].apply(lambda x: [float(val) for _, val in enumerate(x.split())])
#print(len(df_index_lemmas.embeddings.iloc[0]))
```

```python
kept = df_info_objectives[
    (df_info_objectives["kept_in_training"]) |
    (
        df_info_objectives["kept_in_inference"] & 
        df_info_objectives[["kept_by_s3_small", "kept_by_s3_large"]].any(axis=1)
    )
].place_id.values.tolist()
print(len(kept))

ids_to_augment = list(set(kept))
print(len(ids_to_augment))

# keep only one instance of the metadata in df_info_objectives based on "place_id"
df_info_objectives = df_info_objectives.drop_duplicates(subset="place_id")
```

```python
# get dictionary with place_id : raw_text
dict_transform = df_info_objectives.loc[df_info_objectives.place_id.isin(ids_to_augment)][["place_id","raw_text"]].set_index("place_id").to_dict()["raw_text"]
dict_transform
```

```python
# modify the field "title" to df_info_objectives["raw_text"] in the df_index_lemmas if the place_id is in ids_to_augment
df_index_lemmas["title_origin"] = "metadata"

# substitute the title with the raw_text in the df_index_lemmas based on the dict_transform dict
df_index_lemmas["title"] = df_index_lemmas["place_id"].map(dict_transform)
# set df_index_lemmas["title_origin"] to "augmented" if the title was substituted
df_index_lemmas.loc[df_index_lemmas["title"].notnull(), "title_origin"] = "augmented"
```

```python
# find rows where "date" is nan or "nan" and input the current date in the format "YYYY-MM-DD"
import time
df_index_lemmas['date'] = df_index_lemmas['date'].replace("nan", time.strftime("%Y-%m-%d"))
# replace also date that do not follow a correct format. E.g., "0024-02-25" should be "2024-02-25", "0001-01-03" should be "2001-01-03"
def fix_date_format(date):
    parts = date.split("-")
    year = int(parts[0])
    
    # If the year is less than 100 (assuming incorrect formatting), fix it
    if year < 100:
        year += 2000  # Adjusting the year appropriately
        return f"{year}-{parts[1]}-{parts[2]}"
    
    return date

# Apply the function to fix dates
df_index_lemmas['date'] = df_index_lemmas['date'].apply(fix_date_format)

df_index_lemmas
```

```python
df_index_lemmas["date"] = df_index_lemmas["date"].apply(lambda x: x if not x.endswith("'") else x[:-1])
df_index_lemmas[df_index_lemmas['date'].str.endswith("'", na=False)]
```
